### Extracting Category and Subcategory from given Category

```python
"""Here extracted the Actual Category and Subcategory and
assigning them as a new feature of the dataset"""
sales_data['Product Category'] = sales_data['category'].apply(lambda x: x.split('|')[0])
sales_data['Sub Category'] = sales_data['category'].apply(lambda x: x.split('|')[-1])
```

```python
# Finding review count per product and assigning to a new column
sales_data['Reveiw Count'] = sales_data['review_title'].apply(lambda x: len(x.split(',')))
```

### Deleting Unnecessary Columns

```python
# Deleting the columns that don't have any significance
sales_data.drop(columns=['category','img_link','product_link','review_id'],inplace=True)
```

```python
"""Seeing the dataset after adding curated category subcategory 
and rating count"""
sales_data.head()
```

### Data Cleaning

```python
sales_data.dropna(inplace=True)
print(sales_data.isnull().sum())
```

```python
import re
def data_cleaning(data):
    data = str(data)
    data = re.sub(r'[^0-9.]',"",data)
    if data == '':
        return np.nan
    return float(data)
```

```python
sales_data.columns
```

```python
sales_data['actual_price'] = sales_data['actual_price'].apply(lambda x: data_cleaning(x))
sales_data['discounted_price'] = sales_data['discounted_price'].apply(lambda x: data_cleaning(x))
sales_data['rating'] = sales_data['rating'].apply(lambda x: data_cleaning(x))
sales_data['rating_count'] = sales_data['rating_count'].apply(lambda x: data_cleaning(x)).astype(int)
sales_data['discount_percentage'] = sales_data['discount_percentage'].apply(lambda x: data_cleaning(x))
```

```python
sales_data.head()
```

```python
sales_data.isnull().sum()
```

```python
sales_data['rating'].fillna(sales_data['rating'].mean(),inplace=True)
print(sales_data.isnull().sum())
```

## Visualization

```python
#Pair plot of 'actual price', 'discounted price', 'discount percentage'
graph=sns.pairplot(data=sales_data[['actual_price','discounted_price','discount_percentage']])
graph.fig.suptitle("Pair plot of 'actual price', 'discounted price', 'discount percentage'")
plt.show()
```

```python
# Mean rating for each category
mean_rating = sales_data.groupby('Product Category')['rating'].mean()
plt.figure(figsize=(10,5))
plt.bar(mean_rating.index,mean_rating)
plt.title('Mean rating for each category')
plt.xlabel("Product Category")
plt.ylabel("Mean Rating")
plt.xticks(rotation=45)
plt.show()
```

```python
# Mean of price for each category

mean_rated_sales = sales_data.groupby('Product Category')['discounted_price'].mean()
plt.figure(figsize=(10,5))
plt.barh(mean_rated_sales.index,mean_rated_sales)
plt.title('Mean of rated sales for each category')
plt.xlabel("Product Category")
plt.ylabel("Total Sales")
plt.xticks()
plt.show()
```

```python
# showing the category wise rated sales percentage
sales_data['RatedSales'] = sales_data['rating_count']*sales_data['discounted_price']
total_rated_sales = sales_data.groupby('Product Category')['RatedSales'].sum()
total_rated_sales_sum = sales_data['RatedSales'].sum()
percentage_sales = (total_rated_sales/total_rated_sales_sum) *100
plt.pie(percentage_sales,labels=total_rated_sales.index,autopct = "%.2f%%")
plt.title('Category wise sales percentage by sales price')
plt.show()
```

```python
# Number of rated sales
total_rated_sales = sales_data.groupby('Product Category')['rating_count'].sum()
total_rated_sales_sum = sales_data['rating_count'].sum()
percentage_rating = (total_rated_sales/total_rated_sales_sum) *100
plt.pie(percentage_rating,labels=total_rated_sales.index,autopct = "%.2f%%")
plt.title('Category wise Number of rated sales')
plt.show()
```

```python
# rating frequency for top 2 mostly saled category

saled_category = sales_data.groupby('Product Category')['rating_count'].sum().sort_values()
saled_categories = list(saled_category.index)
top_most_saled_categories = saled_categories[-2:]

filtered_sales_1 = sales_data[sales_data['Product Category']==top_most_saled_categories[0]]
filtered_sales_2 = sales_data[sales_data['Product Category']==top_most_saled_categories[1]]

plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
sns.histplot(data=filtered_sales_1, x="rating", bins=10, kde=True, color="green")
plt.title(f'Rating frequency for mostly saled category: {top_most_saled_categories[1]}')

plt.subplot(1,2,2)
sns.histplot(data=filtered_sales_2, x="rating", bins=10, kde=True, color="green")
plt.title(f'\t {top_most_saled_categories[0]}')
plt.show()
```

```python
# Box plot of rating for each category
plt.figure(figsize=(10,5))
sns.boxplot(x="Product Category", y="rating", data=sales_data, palette="Set3", linewidth=1, fliersize=5)
plt.title('Rating for each Category')
plt.xticks(rotation = 45)
plt.show()
```
